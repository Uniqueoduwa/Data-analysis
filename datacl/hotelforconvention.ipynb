```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
listings= pd.read_csv ("listings.csv")
```

```python
listings
```

```python
listings= listings.drop(columns =['longitude', 'latitude','last_review', 'reviews_per_month',  'calculated_host_listings_count','availability_365',])
                        
```

```python
listings_conv =listings.loc [listings ['price']== 150 ,:]
```

```python
listings_conv= listings_conv.loc [listings_conv ['neighbourhood_group']== 'Brooklyn' ,:]
```

```python
listings_conv
```

```python
listings_conv= listings_conv.loc [listings_conv ['room_type']== 'Private room' ,:]
```
